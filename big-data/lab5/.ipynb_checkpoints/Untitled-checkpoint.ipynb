```python
from sklift.datasets import fetch_hillstrom

data = fetch_hillstrom(target_col='visit')
data.keys()
```

```python
X = data['data']
t = data['treatment']
y = data['target']
X.head()
t.value_counts()
y.value_counts()
```

```python
print("Treatment distribution:")
print(t.value_counts(normalize=True))

print("\nTarget distribution:")
print(y.value_counts(normalize=True))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, t_train, t_test, y_train, y_test = train_test_split(
    X, t, y,
    test_size=0.3,
    random_state=42,
    stratify=t  # важно!
)
```

```python
cat_cols = X.select_dtypes(include=['object']).columns
cat_cols
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

encoder.fit(X_train[cat_cols])

X_train_cat = encoder.transform(X_train[cat_cols])
X_test_cat = encoder.transform(X_test[cat_cols])
num_cols = X.select_dtypes(exclude=['object']).columns

X_train_num = X_train[num_cols].values
X_test_num = X_test[num_cols].values
import numpy as np

X_train_final = np.hstack([X_train_num, X_train_cat])
X_test_final = np.hstack([X_test_num, X_test_cat])
```

```python
from sklift.models import TwoModels
from sklearn.ensemble import RandomForestClassifier

t_learner = TwoModels(
    estimator_trmnt=RandomForestClassifier(n_estimators=100, random_state=42),
    estimator_ctrl=RandomForestClassifier(n_estimators=100, random_state=42)
)

t_learner.fit(X_train_final, y_train, t_train)
```

```python
uplift_t = t_learner.predict(X_test_final)
print(uplift_t.shape)
print(uplift_t[:5])
```

```python
from sklift.models import ClassTransformation

ct = ClassTransformation(
    estimator=RandomForestClassifier(n_estimators=100, random_state=42)
)

ct.fit(X_train_final, y_train, t_train)
```

```python
uplift_ct = ct.predict(X_test_final)
print(uplift_ct.shape)
print(uplift_ct[:5])
```

```python
from sklift.metrics import uplift_at_k

print("S-learner:", uplift_at_k(y_test, uplift_s, t_test, strategy='by_group', k=0.3))
print("T-learner:", uplift_at_k(y_test, uplift_t, t_test, strategy='by_group', k=0.3))
print("ClassTransformation:", uplift_at_k(y_test, uplift_ct, t_test, strategy='by_group', k=0.3))

import matplotlib.pyplot as plt

models = ["S", "T", "CT"]
scores = [
    uplift_at_k(y_test, uplift_s, t_test, strategy='by_group', k=0.3),
    uplift_at_k(y_test, uplift_t, t_test, strategy='by_group', k=0.3),
    uplift_at_k(y_test, uplift_ct, t_test, strategy='by_group', k=0.3),
]

plt.bar(models, scores)
plt.title("Uplift@30% comparison")
plt.show()
```
