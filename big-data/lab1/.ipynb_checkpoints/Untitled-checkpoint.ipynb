```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)

n = 5000

workclass = [
"Private","Self-emp-not-inc","Self-emp-inc",
"Federal-gov","Local-gov","State-gov"
]

education = [
"Bachelors","Some-college","11th","HS-grad","Prof-school",
"Assoc-acdm","Assoc-voc","Masters","Doctorate"
]

marital_status = [
"Married-civ-spouse","Divorced","Never-married",
"Separated","Widowed","Married-spouse-absent"
]

occupation = [
"Tech-support","Craft-repair","Other-service","Sales",
"Exec-managerial","Prof-specialty","Adm-clerical"
]

relationship = [
"Wife","Own-child","Husband","Not-in-family","Unmarried"
]

race = [
"White","Asian-Pac-Islander","Amer-Indian-Eskimo",
"Other","Black"
]

sex = ["Female","Male"]

countries = [
"United-States","Germany","Canada","India",
"England","France","Mexico"
]

salary = ["<=50K",">50K"]

df = pd.DataFrame({

"age": np.random.randint(18,70,n),

"workclass": np.random.choice(workclass,n),

"fnlwgt": np.random.randint(10000,100000,n),

"education": np.random.choice(education,n),

"education-num": np.random.randint(1,16,n),

"marital-status": np.random.choice(marital_status,n),

"occupation": np.random.choice(occupation,n),

"relationship": np.random.choice(relationship,n),

"race": np.random.choice(race,n),

"sex": np.random.choice(sex,n),

"capital-gain": np.random.choice([0,0,0,0,5000,10000],n),

"capital-loss": np.random.choice([0,0,0,2000],n),

"hours-per-week": np.random.randint(20,60,n),

"native-country": np.random.choice(countries,n),

"salary": np.random.choice(salary,n)

})

df.head()
```

```python
men = df[df["sex"] == "Male"]

married = men["marital-status"].str.startswith("Married")

men.groupby(married)["salary"].apply(lambda x: (x==">50K").mean())
```

```python
country_stats = df.groupby("native-country").agg(
    total=("salary","count"),
    high_income=("salary", lambda x: (x==">50K").sum())
)

country_stats = country_stats[country_stats["total"] >= 50]

country_stats["share"] = country_stats["high_income"] / country_stats["total"]

country_stats.sort_values("share", ascending=False).head(1)
```
